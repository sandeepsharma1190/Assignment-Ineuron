### 1. To what does a relative path refer?

Answer : - Relative paths are relative to the current working directory.

### 2. What is the first step on an absolute path?
Answer: - An Absolute path of any directory always starts with a slash (/) or C:\ representing the directory root.

### 3. On Windows, what does Path('C:/Users') / 'Al' evaluate to?

Answer : - On Windows, it evaluates to a WindowsPath('C:/Users/Al').

### 4. On Windows, what does 'C:/Users' / 'Al' evaluate to?
Answer : - We can’t use the / operator to join two strings. The expression 'C:/Users' / 'Al' will results in an error.

### 5. What do the functions os.getcwd() and os.chdir() do?
Answer: - The os.getcwd() function returns the current working directory. The os.chdir() function changes the current working directory.

### 6. What is the difference between the. and.. folders?
Answer: - The . folder is the current folder, and .. is the parent folder.

### 7. Which part of C:baconeggsspam.txt is the dir name and which part is the base name?
Answer: - C:\bacon\eggs is the dir name, while spam.txt is the base name.

### 9. What happens if you open existing file in write mode?
Answer : - in write mode, given text will overwrite all texts in file. old text will be erased.

### 10. How do you tell the difference between read() and readlines()?

Answer: - The read() method will return the file's entire content as a single string value. 

The readlines() method will return a list of strings, where each string is a line from the file's content.
