### 9. Create a programme that prints. If 1 is stored in spam, prints Hello; if 2 is stored in spam, prints Howdy; and if 3 is stored in spam, prints Salutations! if there's something else in spam.

```python
spam = int(input('Enter a number: '))
if spam == 1:
    print ('Hello')
elif spam == 2:
    print ('Howdy')
elif spam == 3:
    print ('Salutations')
else:
    print ('Spam')
```

### 11. How can you tell the difference between break and continue?
Answer - We use break when we want some condition to met and break the loop. 'break' stops the continuation of loop.
We use continue when we want to terminate the current iteration and resumes the next iteration of the loop. ‘continue' do not stops the continuation of loop.


### 12. In a for loop, what is the difference between range(10), range(0, 10), and range(0, 10, 1)?
Answer - They all do the same thing. 

range(10) – means values from 0-9, 10 will be excluded as an upper bound value. By default 0 is start and 10 end.

range(0,10) - means values from 0-9, 10 will be excluded as an upper bound value. 0 is start and 10 is end.

range(0, 10, 1) - means values from 0-9, 10 will be excluded as an upper bound value with 1 iteration. 0 means start, 10 means end (upperbound) and 1 will be step(s).
