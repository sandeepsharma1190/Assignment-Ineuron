### 1.	Write a Python Program to Find LCM?

```python
x = int(input('Enter first num: '))
y = int(input('Enter second num: '))

if x > y:
    num = x
else:
    num = y
    
while True:
    if (num % x == 0 and num % y == 0):
        print ('LCM of {0} and {1} = {2}'.format(x,y,num))
        break
    num = num+1
```

### 3.	Write a Python Program to Convert Decimal to Binary, Octal and Hexadecimal?

```python
num1 = int(input('Enter a num for its conversion: '))
print ('')
print ('Binary value for', num1, 'is', bin(num1))
print ('Octal value for', num1, 'is', oct(num1))
print ('Hexadecimal value for', num1, 'is', hex(num1))
```

### 4.	Write a Python Program To Find ASCII value of a character?

```python
data1 = input('Enter a character for its ASCII conversion: ')
print ('ASCII value for', data1, 'is', ord(data1))
```

### 5.	Write a Python Program to Make a Simple Calculator with 4 basic mathematical operations?

```python
def add (a,b):
    return a+b

def subtraction (a,b):
    return a-b

def multiplication (a,b):
    return a*b

def divide (a,b):
    return a/b

print ('Select any below mentioned operation')
print ('1 for Addition')
print ('2 for Subtraction')
print ('3 for Multiplication')
print ('4 for Divide')

while True:
    choice = input('Enter your choice: ')
    if choice in ('1', '2', '3', '4'):
        num1 = float(input('Enter first number: '))
        num2 = float(input('Enter second number: '))
        
        if choice == '1':
            print ('Addition for', num1, 'and', num2, 'is', add(num1,num2))

        elif choice == '2':
            print ('Subtraction for', num1, 'and', num2, 'is', subtraction(num1+num2))

        elif choice == '3':
            print ('Multiplication for', num1, 'and', num2, 'is', multiplication(num1,num2))
            
        elif choice == '4':
            print ('Divide for', num1, 'and', num2, 'is', divide(num1,num2))
        break
    else:
        print ('Invalid Input')
```
