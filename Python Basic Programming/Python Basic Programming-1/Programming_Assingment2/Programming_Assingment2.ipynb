### 1.	Write a Python program to convert kilometers to miles?

```python
km = float(input('Enter kilometers: '))
km_to_mile = 0.621371
mile = km*km_to_mile
print (km, 'Kilometer', 'is equal to', mile, 'miles')
```

### 2.	Write a Python program to convert Celsius to Fahrenheit?

```python
celsius = float(input('Enter Celsius temprature: '))
def_temp = 32
fahrenheit = (celsius*1.8) + def_temp
print (celsius, 'celsius', 'is equal to', fahrenheit, 'fahrenheit')
```

### 4.	Write a Python program to solve quadratic equation?

```python
import math 
def equationroots( a, b, c): 
    dis = b * b - 4 * a * c 
    sqrt_val = math.sqrt(abs(dis)) 

    if dis > 0: 
        print("real and different roots") 
        print((-b + sqrt_val)/(2 * a)) 
        print((-b - sqrt_val)/(2 * a)) 
    elif dis == 0: 
        print("real and same roots") 
        print(-b / (2 * a)) 
    else:
        print("Complex Roots") 
        print(- b / (2 * a), " + i", sqrt_val) 
        print(- b / (2 * a), " - i", sqrt_val) 
  
a = 5
b = 10
c = 15

if a == 0:
    print("Input correct quadratic equation") 
else:
    equationroots(a, b, c)
```

### 5.	Write a Python program to swap two variables without temp variable?

```python
a,b = 'Hello', 'Python'
print ('Value of a is:', a)
print ('Value of b is:', b)
print ('-----------------------')
b,a = a,b
print ('Updated value of a is:', a)
print ('Updated value of b is:', b)
```
