### 3. What is the most significant distinction between a dictionary and a list?
Answer : - Dictionary contains two values ({‘foo’ :42 }, ‘foo’ is the item, 42 is the value and “’ foo’ :42” is an item.), with no index values. while list contains only single value with index information. 
Data in dictionary can be unordered, however data in list is ordered.
